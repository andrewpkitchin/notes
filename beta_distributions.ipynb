```python
import random as rand

random_prob = rand.random()
```

```python
# prompt: I'd like to create some random binomial trial data in a list.

import numpy as np
data = np.random.binomial(1, random_prob, 25)
print(data)
```

```python
hypothesis = sum(data)/len(data)

print(hypothesis)

print(sum(data), len(data))
```

```python
import matplotlib.pyplot as plt
from scipy.stats import beta

# Parameters of the beta distribution
a = sum(data)
b = len(data) - sum(data)

# Generate x values for plotting
x = np.linspace(0, 1, 100)

# Calculate the corresponding probability density function (PDF) values
y = beta.pdf(x, a, b)

# Plot the beta distribution
plt.plot(x, y)
plt.title('Beta Distribution (a={}, b={})'.format(a, b))
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.show()
```
